## 1. 바다ocean

```python
driver = webdriver.Chrome()
driver.maximize_window()
driver.get('https://onthelook.co.kr/searchV4?q=&p=1&v=POST&f={%22tpoTagId%22:36}')
time.sleep(5)

# Scroll to the bottom of the page to load all images
last_height = driver.execute_script("return document.body.scrollHeight")
while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)  # Adjust the sleep time as needed
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

# Find all images after scrolling
image_list = driver.find_elements(By.CSS_SELECTOR, '#root > div.sc-kpOJdX.hShvMP > div > main > div.sc-gacfCG.kJoPTh > div.sc-cqPOvA.hmgoEW > div > a > div > img')
print("Total images found:", len(image_list))

# Create a directory to save the images
if not os.path.exists('ocean'):
    os.makedirs('ocean')

# Download and save each image
for i, image in enumerate(image_list):
    src = image.get_attribute('src')
    response = requests.get(src)
    
    # Extract the image filename from the URL
    filename = os.path.join('ocean', f"image_{i + 1}.jpg")
    
    # Save the image to the specified directory
    with open(filename, 'wb') as file:
        file.write(response.content)

    print(f"Image {i + 1} saved as {filename}")

driver.quit()
```
